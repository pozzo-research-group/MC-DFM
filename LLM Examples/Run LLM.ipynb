## RAG/LLM MC-DFM
This notebook demonstrates a retrieval-augmented language model (RAG/LLM) workflow for generating real-space structural models and computing their small-angle scattering curves using the MC-DFM method. The user provides a natural-language description of the desired structure, the LLM generates Python code to build the corresponding model, and the validated MC-DFM routines are then used to simulate the scattering intensity. The notebook is used to instruct the workflow on what kind of structrues to simulate. The output is a folder containing a python file called 'generated_script.py' in the that is automatically saved in the **save_dir** directory. 

To use the workflow, follow these steps:
1. This workflow uses an LLM through the AtomGPT API . To use it, first create a free account at https://atomgpt.org/. Then navigate to Settings → Account → Show API key and copy your API key. Assign the copied value to the **api_key** variable before running the notebook.
2. Enter the LLM model to be used. If you dont have a preference, leave **model = "openai/gpt-oss-20b"**
3. Enter the directory where the results are saved by assigning it to **save_dir**
4. Enter the **polydispersity_mode** as either **'weighted_sum** or **'distribution'**. This determines how polydispersity is modeled. 
5. Enter the instructions on what kind of structures you wish to be simulated by assigning it to **instructions**

```python
api_key = 'Enter_your_API_Key_here'
model="openai/gpt-oss-20b"
save_dir = "../sas_llm_results/"
polydispersity_mode = "weighted_sum" #or "weighted_sum or distribution"
instructions = 'Simulate the scattering of a square-based pyramid that has an edge length of 10 nm and a height of 10 nm.'
```

The next code block runs the workflow and the results will be stored in a folder in the **save_dir** directory. This can take up to a minute long depending on how complex the instrctions are. 

```python
await use_llm(api_key, model, polydispersity_mode, instructions, save_dir)
```

The next code block executes the python file created by the workflow named 'generated_script.py'. You can also copy and paste the code from the script in this notebook and it should run.

```python
filenames = os.listdir(save_dir)
%run {save_dir + filenames[-1] + '/generated_script.py'}
```

## Troubleshooting Errors 
1. If https://atomgpt.org/ is not working, this workflow will also not work. Check if the website is up and running. 
2. Sometimes the output python code contains coding/syntax errors. When this happens, try re-running the workflow with the same instructions and the output should run most of the time. 
3. Always check the plots showing the 3D and 2D representations of the generated structure to confirm whether the LLM correctly generated your desired structure. 
4. Try to avoid giving extremely long instructions such as generating complex structures and then fitting. Instead, split it into two separate prompts. 
5. The LLM treats each prompt independently and will forget any previous information that you give it.
